# Skewness Method

```python
skewness = new_dataset.skew()


abs(skewness)

# setting up the threshold
skewness_threshold = 0.5


symmetrical_features = skewness[abs(skewness) < skewness_threshold].index
skewed_features = skewness[abs(skewness) >= skewness_threshold].index


print('SYMMETRICAL DISTRIBUTION :')
symmetrical_data = new_dataset[symmetrical_features]
print(symmetrical_features)

print('SKEWED DISTRIBUTION :')
skewed_data = new_dataset[skewed_features]
print(skewed_features)
```

```python
skewed_data
```

### Symmetrizing distributions with skewed distributions

```python
#import the liabrary
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer

# Initialize the PowerTransformer (It brings skewed distributions closer to the Gaussian (normal) distribution.)
power_transformer = PowerTransformer()

# Fit and transform the data using the PowerTransformer
power_transformed = pd.DataFrame(power_transformer.fit_transform(skewed_data))
power_transformed.columns = skewed_data.columns
```

```python
# Reset the index to the default integer index
symmetrical_data.reset_index(drop=True, inplace=True)
```

```python
symmetrical_data
```

```python
#Combining previous data with transformed data
tranformed_data = pd.concat([symmetrical_data, power_transformed], axis=1)
```

```python
tranformed_data
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(tranformed_data))
scaled_test_data = pd.DataFrame(scaler.fit_transform(test_data))
scaled_data.columns = tranformed_data.columns
scaled_test_data.columns = test_data.columns

scaled_data
```

```python
scaled_test_data
```

# PCA (Principal Component Analysis)

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


scaled_test_data = scaled_test_data[scaled_data.columns]


pca = PCA()
pca.fit(scaled_data)

# Cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Elbow plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Elbow Plot')
plt.grid()
plt.show()


n_components = 10
pca = PCA(n_components=n_components)


pca.fit(scaled_data)


transformed_data_pca = pca.transform(scaled_data)
transformed_test_data_pca = pca.transform(scaled_test_data)


explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio (for each component):", explained_variance)
```

# Model Training

```python
x = transformed_data_pca
y = new_dataset['Appliances']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures



x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=3)
poly_features = PolynomialFeatures(degree=2)

# Transforming data for polynomial usable
X_train_poly = poly_features.fit_transform(x_train)
X_test_poly = poly_features.transform(x_test)
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from joblib import dump, load


rf_regr = RandomForestRegressor(n_estimators=20, random_state=42)


rf_regr.fit(X_train_poly, y_train)


train_predictions_rf = rf_regr.predict(X_train_poly)
test_predictions_rf = rf_regr.predict(X_test_poly)


train_mse_rf = mean_squared_error(y_train, train_predictions_rf)
test_mse_rf = mean_squared_error(y_test, test_predictions_rf)

train_r2_rf = r2_score(y_train, train_predictions_rf)
test_r2_rf = r2_score(y_test, test_predictions_rf)

print("Train MSE:", train_mse_rf)
print("Test MSE:", test_mse_rf)
print("Train R2:", train_r2_rf)
print("Test R2:", test_r2_rf)

#Save Model
dump(rf_regr, "random_forest_model.joblib")

```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
regr_gb = GradientBoostingRegressor(random_state=42).fit(x_train, y_train)
train_score = regr_gb.score(x_train, y_train)
test_score = regr_gb.score(x_test, y_test)
#predicting the value
train_predictions = regr_gb.predict(x_train)
test_predictions = regr_gb.predict(x_test)
print("Train score:" ,train_score,"Test score:" ,test_score)
train_r2 = r2_score((y_train),(train_predictions))
test_r2 = r2_score((y_test),(test_predictions))
print("Train R2 :" ,train_r2, "Test R2 :" ,test_r2)
MSE  = mean_squared_error((y_test),(test_predictions))
print("Test MSE :" , MSE)

dump(regr_gb, "gr_boost")
```

## XGBoost

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import joblib


# Define the parameter grid
param_grid = {
    'n_estimators': [100], 
    'learning_rate': [0.01, 0.1, 0.2],  # Learning rate
}

xgb_regr = xgb.XGBRegressor(random_state=3)

grid_search = GridSearchCV(xgb_regr, param_grid, cv=5, scoring='neg_mean_squared_error')

grid_search.fit(X_train_poly, y_train)

# Get the best hyperparameters from the grid search
best_params = grid_search.best_params_

# Create an XGBoost Regressor model with the best hyperparameters
best_xgb_regr = xgb.XGBRegressor(**best_params, random_state=3)

# Fit the best XGBoost model to the training data
best_xgb_regr.fit(X_train_poly, y_train)


train_predictions_xgb = best_xgb_regr.predict(X_train_poly)
test_predictions_xgb = best_xgb_regr.predict(X_test_poly)

train_mse_xgb = mean_squared_error(y_train, train_predictions_xgb)
test_mse_xgb = mean_squared_error(y_test, test_predictions_xgb)

train_r2_xgb = r2_score(y_train, train_predictions_xgb)
test_r2_xgb = r2_score(y_test, test_predictions_xgb)

print("Best Hyperparameters:", best_params)
print("Train MSE:", train_mse_xgb)
print("Test MSE:", test_mse_xgb)
print("Train R-:", train_r2_xgb)
print("Test R2:", test_r2_xgb)

dump(best_xgb_regr, "xgb_model.joblib")
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression().fit(x_train, y_train)
train_score = regr.score(x_train, y_train)
test_score = regr.score(x_test, y_test)
#predicting the value
train_predictions = regr.predict(x_train)
test_predictions = regr.predict(x_test)
print("Train score:" ,train_score,"Test score:" ,test_score)
train_r2 = r2_score((y_train),(train_predictions))
test_r2 = r2_score((y_test),(test_predictions))
print("Train R2 :" ,train_r2, "Test R2 :" ,test_r2)
MSE  = mean_squared_error((y_test),(test_predictions))
print("Test MSE :" , MSE)

dump(regr, "linear_reg.joblib")
```
